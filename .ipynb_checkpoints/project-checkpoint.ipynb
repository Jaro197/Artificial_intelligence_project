## Read in my data
The data comes from the official api of the cryptocurrency BitBay exchange https://docs.bitbay.net/reference, where you can see the full specification. These are 52 files, where each file contains hourly snapshots of the stock exchange. 

```python
#specify the name of the data folder
director_name = 'btc_data/'
filelist = os.listdir(director_name)

# read in json file data
df_list = [pd.read_json(director_name + file) for file in filelist]
```

## Process my data

```python
# set the time you want to predict rate
in_hours = 2

# add futreRate column to be predicted
for i in range(len(df_list)):
    df_list[i]['futureRate'] = df_list[i].rate[in_hours:].reset_index(drop=True)
    
# merge all files in to one DataFrame
df = pd.concat(df_list)

# remove all 'null' from timeUnix column
df = df[df['timeUnix'] != "null"]

# remove all 'null' from futureRate column
df = df[df['futureRate'] != "null"]

# remove unnecessary column
df.drop(columns=['code'], inplace=True)

# remove Nan
df = df.dropna()

# set types
df['timeUnix'] = df['timeUnix'].astype(int)
df[['highestBid', 'lowestAsk', 'rate', 'previousRate', 'highest24', 'lowest24', 'volumen24', 'average24', 'futureRate']] = \
df[['highestBid', 'lowestAsk', 'rate', 'previousRate', 'highest24', 'lowest24', 'volumen24', 'average24', 'futureRate']].astype(float)

# look at data type again
display(df.dtypes)

# sort rows by date
df.sort_values(by=['timeUnix'], inplace=True)

# reset index
df = df.reset_index(drop=True)

# check if date is uniqe
print(df['timeUnix'].is_unique)

# change timeUnix to datetime
df['timeUnix'] = df['timeUnix'].apply(lambda x: datetime.fromtimestamp(x / 1000.))
```

```python
def hour_rounder(t):
    # Rounds to nearest hour by adding a timedelta hour if minute >= 30
    return (t.replace(second=0, microsecond=0, minute=0, hour=t.hour)
               +timedelta(hours=t.minute//30))
```

```python
# take 'ra' - position rate
for i in range(1, 11):
    df[f"sell_{i}"] = eval(f"df.apply(lambda row: float(row.sell_{i}['ra']), axis=1)")
    df[f"buy_{i}"] = eval(f"df.apply(lambda row: float(row.buy_{i}['ra']), axis=1)")    
    
# create mean and std of all sell/buy
df['meanSell10'] = round(np.mean(df.loc[:, 'sell_1':'sell_10'], axis=1), 2)
df['stdSell10'] = round(np.std(df.loc[:, 'sell_1':'sell_10'], axis=1), 2)
df['meanBuy10'] = round(np.mean(df.loc[:, 'buy_1':'buy_10'], axis=1), 2)
df['stdBuy10'] = round(np.std(df.loc[:, 'buy_1':'buy_10'], axis=1), 2)

# remove unnecessary column
df.drop(columns=['sell_{}'.format(i) for i in range(1, 11)], inplace=True)
df.drop(columns=['buy_{}'.format(i) for i in range(1, 11)], inplace=True)

# rounds to nearest hour
df['timeUnix'] = df.apply(lambda row: hour_rounder(row.timeUnix), axis=1)

# look at data
df.head()
```

## Regression: Ridge, Lasso and Elastic Net
Let's try to predict the future rate by using classic machine learning methods.

```python
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
```

```python
# create dataset without timeUnix
dataset = df.loc[:, df.columns != 'timeUnix'].values
dataset = dataset.astype('float32')

# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)

# split into train and test sets
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
print(len(train), len(test))
X_train, y_train = np.delete(train, [8], 1), train[:,8]
X_test, y_test = np.delete(test, [8], 1), test[:,8]
```

```python
seed = 123
kfold = model_selection.KFold(n_splits=4, random_state=seed)
```

```python
# ElasticNet
grid_1 = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), ElasticNet(alpha=1, max_iter=10000)),
                    param_grid={'polynomialfeatures__degree': [1, 2, 3, 4],
                    'elasticnet__alpha': [0.001, 0.01, 0.1, 1., 10., 100., 1000.]},
                    cv=kfold,
                    refit=True,
                    n_jobs=-1)
grid_1.fit(X_train, y_train)
print(grid_1.best_params_)

# plot prediction
plt.plot(y_test);
plt.plot(grid_1.predict(X_test));
```

```python
# Lasso
grid_2 = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), Lasso(alpha=1)),
                    param_grid={'polynomialfeatures__degree': [1, 2, 3, 4, 5],
                    'lasso__alpha': [0.001, 0.01, 0.1, 1., 10., 100., 1000.]},
                    cv=kfold,
                    refit=True,
                    n_jobs=-1)
grid_2.fit(X_train, y_train)
print(grid_2.best_params_)

# plot prediction
plt.plot(y_test);
plt.plot(grid_2.predict(X_test));
```

```python
#Ridge
grid_3 = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), Ridge(alpha=0.1)),
                   param_grid={'polynomialfeatures__degree': [1, 2, 3, 4, 5],
                               'ridge__alpha': [0.001, 0.01, 0.1, 1, 10, 100, 1000]},
                   cv=kfold,
                   refit=True,
                   n_jobs=-1)
grid_3.fit(X_train, y_train)
print(grid_3.best_params_)

# plot prediction
plt.plot(y_test);
plt.plot(grid_1.predict(X_test));
```

## MLPRegressor

```python
from sklearn.neural_network import MLPRegressor
```

```python
param_grid = {
            'hidden_layer_sizes': [(200, 100, 10), (100, 50, 10), (150, 100, 50)],
            'max_iter': [5000, 10000],
            'batch_size': [10, 20, 30, 40],
            'learning_rate_init': [0.0001, 0.001, 0.01],
            'activation': ['than', 'relu', 'logistic', 'sigmoid']        
}

grid_4 = GridSearchCV(MLPRegressor(learning_rate="adaptive", solver='adam'), param_grid, cv=5, return_train_score=True, n_jobs=-1)
grid_4.fit(X_train, y_train)
print(grid_4.best_params_)

# plot prediction
plt.plot(y_test);
plt.plot(grid_4.predict(X_test));
```

## Keras neural network

```python
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping
import keras
```

```python
early_stopping = EarlyStopping(patience=10, monitor="val_loss")

grid_5 = keras.models.Sequential([
    keras.layers.Dense(250, activation="relu", input_shape=X_train.shape[1:]),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(150, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(50, activation="relu"),
    keras.layers.Dense(1)
])
grid_5.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
history = grid_5.fit(X_train, y_train, epochs=50, validation_split=0.1, callbacks=[early_stopping],
           use_multiprocessing=True)

# plot prediction
plt.plot(y_test);
plt.plot(grid_5.predict(X_test));
```

```python
plt.plot(pd.DataFrame(history.history))
plt.grid(True)
plt.show()
```

## LSTM
Let's try to predict the future rate by using LSTM.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Embedding, SimpleRNN, LSTM, Bidirectional
from keras.callbacks import History
```

```python
#reshape input to be [samples, time steps, features]
trainX = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
testX = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
# create and fit the LSTM network
history_2 = History()

grid_7 = Sequential()
grid_7.add(LSTM(4, input_shape=(1, 12)))
grid_7.add(Dense(1))
grid_7.compile(loss='mean_squared_error', optimizer='adam', metrics=["accuracy"])
grid_7.fit(trainX, y_train, epochs=10, batch_size=1, validation_split=0.1, callbacks=[history_2])

# plot prediction
plt.plot(y_test);
plt.plot(grid_7.predict(testX));
```

```python
pd.DataFrame(history_2.history).plot(figsize=(8, 5))
plt.grid(True)
plt.show()
```

## Comparison of all models

```python
models = []
models.append(('ElasticNet', grid_1.best_estimator_))
models.append(('Lasso', grid_2.best_estimator_))
models.append(('Ridge', grid_3.best_estimator_))
models.append(('MLP_R', grid_4.best_estimator_))
models.append(('keras', grid_5))
models.append(('LR', grid_6.best_estimator_))

r2 = []
explained_variance_score = []
median_absolute_error = []
mean_squared_error = []
mean_absolute_error = []
for name, model in models:
    print(name)
    print("R^2: {}".format(metrics.r2_score(y_test, model.predict(X_test)) ))
    print("Explained variance score: {}".format( metrics.explained_variance_score(y_test, model.predict(X_test)) ))
    print("Median absolute error: {}".format( metrics.median_absolute_error(y_test, model.predict(X_test)) ))
    print("Mean squared error: {}".format( metrics.mean_squared_error(y_test, model.predict(X_test)) ))
    print("Mean absolute errors: {}".format(metrics.mean_absolute_error(y_test, model.predict(X_test)) ))
    r2.append(metrics.r2_score(y_test, model.predict(X_test)))
    explained_variance_score.append(metrics.explained_variance_score(y_test, model.predict(X_test)))
    median_absolute_error.append( metrics.median_absolute_error(y_test, model.predict(X_test)))
    mean_squared_error.append(metrics.mean_squared_error(y_test, model.predict(X_test)))
    mean_absolute_error.append(metrics.mean_absolute_error(y_test, model.predict(X_test)))

r2.append(metrics.r2_score(y_test, grid_7.predict(testX)))
explained_variance_score.append(metrics.explained_variance_score(y_test, grid_7.predict(testX)))
median_absolute_error.append( metrics.median_absolute_error(y_test, grid_7.predict(testX)))
mean_squared_error.append(metrics.mean_squared_error(y_test, grid_7.predict(testX)))
mean_absolute_error.append(metrics.mean_absolute_error(y_test, grid_7.predict(testX)))
```

```python
d = {'r2': r2, 
     'explained_variance_score': explained_variance_score, 
     'median_absolute_error': median_absolute_error,
     'mean_squared_error' : mean_squared_error,
     'mean_absolute_error' : mean_absolute_error,
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Method', value=['ElasticNet', 'Lasso', 'Ridge', 'MLPR', 'keras', 'LR', 'LSTM'])
df.sort_values(by=['r2'], ascending=False, ignore_index=True)
```
